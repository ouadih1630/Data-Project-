```python
import pandas as pd
#pd.set_option('display.max_columns',None)
results=pd.read_csv('/content/data.csv')
results
```

```python
results.columns
```

```python
results= results[['description', 'emplacement', 'type', 'Surface m²',
       'Price €', 'nb_piece']]
results
```

```python
results.describe()
```

```python
results["loyer / m²"] = results["Price €"] / results["Surface m²"]
results.describe()
```

```python
results.groupby("emplacement")["loyer / m²"].mean().sort_values(ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (8,4.5)
sns.violinplot(x="nb_piece", y="loyer / m²",data=results)
```

```python
homes = results.groupby(["emplacement", "nb_piece"])["loyer / m²"].mean().to_frame()
homes = homes.reset_index()
homes = homes.pivot("emplacement", "nb_piece", "loyer / m²")
sns.heatmap(homes, annot=True, cmap="YlGnBu")
```
